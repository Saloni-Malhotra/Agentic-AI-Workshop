```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    message: str
```

```python
def greeting_node(state: AgentState) -> AgentState:
    """A simple node that says hello to caller name. It's nice to be polite!"""

    state['message'] = "Hey "+state['message']+"! How are you?"
    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("Pookie",greeting_node)
graph.set_entry_point("Pookie")
graph.set_finish_point("Pookie")

app=graph.compile()
```

```python
from IPython.display import display, Image
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result=app.invoke({"message": "Saloni"})
print(result)
```
